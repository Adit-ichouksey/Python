```python
def even_factorial(n):
    fact=1
    for i in range(2,n+1,2):
        fact=fact*i
    return fact

def odd_factorial(n):
    fact=1
    for i in range(1,n+1,2):
            fact=fact*i
    return fact
# fact=1
n=int(input("Enter a number : "))
if n%2==0:
    print(even_factorial(n))
else:
    print(odd_factorial(n))
```

```python
# calculator nCr
def fact(n):
    fact=1
    for i in range(1,n+1):
        fact=fact*i
    return fact

n=10
r=2
nfact=factorial(n)
rfact=factorial(r)
nrfact=factorial(n-r)
ans= nfact//(rfact*nrfact)
print(ans)
```
