```python
def is_valid_email(email):
    if "@" in email and "." in email:
        at_index = email.index("@")
        dot_index = email.rindex(".")
        return at_index < dot_index
    return False

# Example
print(is_valid_email("example@domain.com"))  # True
print(is_valid_email("example.domain@com"))  # False
```

```python
def reverse_words(sentence):
    words = sentence.split()
    return " ".join(words[::-1])

# Example
print(reverse_words("Hello World from Python"))
```

```python
def find_indices(s, ch):
    return [i for i, c in enumerate(s) if c == ch]

# Example
print(find_indices("programming", "m"))

```

```python
def remove_substring(s, sub):
    return s.replace(sub, "")

# Example
print(remove_substring("This is a test string for testing", "test"))

```
