```python
t = (1,9,1,6,3,4,5,1,1,2,5,6,7,8,9,2)

unique = ()
for i in t:
    if i not in unique:
        unique += (i,)

print(unique)
```

```python
t = (('333','33'),('1416','55'))

new_t=()

for i in t:
    temp =()
    for j in i:
        temp = temp + (int(j),)
        new_t = new_t + (temp,)

print(new_t)
```

```python
import math
# First tuple example
tpl1 = (4, 3, 2, 2, -1, 18)
product1 = math.prod(tpl1)
print("Original tuple:", tpl1)
print("Product - multiplying all the numbers of the said tuple:", product1)


# Second tuple example
tpl2 = (2, 4, 8, 8, 3, 2, 9)
product2 = math.prod(tpl2)
print("\nOriginal tuple:", tpl2)
print("Product - multiplying all the numbers of the said tuple:", product2)
```
