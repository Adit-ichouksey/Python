```python
l1 = [1,2,3,4]
l2 = [3,4,5,6]
l3 = [1,4,6,7]
result = set(l1) | set(l2) | set(l3)
common = set(l1) & set(l2) & set(l3)
S = result - common
print(S)
```

```python
S = {1,2,3}
L =[2,3,4,5]

for i in L:
    if i not in S:
        S.add(i)
    else:
            S.remove(i)
print(S)
```

```python
A = {2,3,5}
B = {5,7,11}
C = {11,13,17}

S = A | B | C
for i in list (S):
    if i> 10:
        S.remove(i)
print(S)
```
