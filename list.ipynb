```python
lst=[[1,30,5,6],[3,4,5],[3,5,7],[10,20,30]]
ans=[]
for i in range(len(lst)):
    total=0
    for j in range(len(lst[i])):
        temp=lst[i].copy()
        if i%2==0:
          for j in range(len(lst[i])):
            if temp[j]%2==0:
                temp[j]-=1
                ans.append(temp)
        else:
           for j in range(len(temp)):
            if temp[j]%2!=0:
                temp[j]+=1
                ans.append(temp)
print(lst)
print(ans)
```

```python
# WAP to create a list of squares of even no and square of odd no 

lst = [6,9,12,15,18,21,24]
even_squares = [x**2 for x in lst if x % 2 == 0]
odd_squares = [x**2 for x in lst if x % 2 != 0]
print("Even squares:", even_squares)
print("Odd squares:", odd_squares)
```
